```python
import pandas as pd

file_path = "IMDB Dataset.csv"

df = pd.read_csv(file_path, engine='python', on_bad_lines='warn')

reviews = df["review"].str.lower().tolist()
sentiments = df["sentiment"].str.lower().tolist()

total_reviews = len(reviews)
positive_reviews = [reviews[i] for i in range(total_reviews) if sentiments[i] == "positive"]
negative_reviews = [reviews[i] for i in range(total_reviews) if sentiments[i] == "negative"]

print("Total reviews:", total_reviews)
```

```python
P_negative = len(negative_reviews) / total_reviews
print("P(Negative):", P_negative)
```

```python
def likelihood(keyword):
    negative_reviews = [
        reviews[i]
        for i in range(len(reviews))
        if sentiments[i] == "negative"
    ]

    count = sum(1 for review in negative_reviews if keyword in review)

    return count / len(negative_reviews)
```

```python
def marginal(keyword):
    count = sum(1 for review in reviews if keyword in review)
    return count / total_reviews
```

```python
def posterior(keyword):
    P_kw_neg = likelihood(keyword)
    P_kw = marginal(keyword)

    if P_kw == 0:
        return 0

    return (P_kw_neg * P_negative) / P_kw
```

```python
keywords = ["liked", "enjoyed", "terrible", "worst"]

results = []

for word in keywords:
    results.append({
        "Keyword"              : word,
        "Prior P(Neg)"         : round(P_negative,       4),
        "Likelihood P(kw|Neg)" : round(likelihood(word), 4),
        "Marginal P(kw)"       : round(marginal(word),   4),
        "Posterior P(Neg|kw)"  : round(posterior(word),  4),
    })

results_df = pd.DataFrame(results)
display(results_df)

```

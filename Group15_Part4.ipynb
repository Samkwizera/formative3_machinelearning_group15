```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize   # Imported as requested


x = np.array([1, 3])
y = np.array([3, 6])
n = len(x)


# INITIAL VALUES
m = -1
b = 1
alpha = 0.1
iterations = 4

# STORAGE FOR PLOTS
m_values = []
b_values = []
error_values = []

# GRADIENT DESCENT LOOP
for i in range(iterations):

    # STEP 1: Compute predictions
    y_pred = m * x + b

    # STEP 2: Compute errors
    errors = y_pred - y

    # STEP 3: Compute gradients
    # Using J = (1/n) * sum(...)^2  -> gradients without 1/n
    dm = np.sum(errors * x)
    db = np.sum(errors)

    # STEP 4: Update parameters
    m = m - alpha * dm
    b = b - alpha * db

    # STEP 5: Compute cost (MSE version used in assignment)
    cost = (1/n) * np.sum(errors**2)

    # Store values
    m_values.append(m)
    b_values.append(b)
    error_values.append(cost)

    # Print each step clearly
    print("Iteration:", i+1)
    print("m =", m)
    print("b =", b)
    print("Cost =", cost)
    print("---------------------")

# FINAL PREDICTIONS
final_predictions = m * x + b

print("Final m:", m)
print("Final b:", b)
print("Final Predictions:", final_predictions)
```

```python
# m and b Over Iterations
plt.figure()
plt.plot(range(1, iterations+1), m_values, marker='o', label='m')
plt.plot(range(1, iterations+1), b_values, marker='o', label='b')
plt.xlabel("Iteration")
plt.ylabel("Parameter Value")
plt.title("Change of m and b Over Iterations")
plt.legend()
plt.show()
```

```python
# Error Over Iterations
plt.figure()
plt.plot(range(1, iterations+1), error_values, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Cost (Error)")
plt.title("Error Reduction Over Iterations")
plt.show()
```
